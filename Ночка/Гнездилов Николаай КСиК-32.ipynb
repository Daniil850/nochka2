```python
#Решение способом обратной матрицы 
import numpy as np
A=np.array([[-1, -3,  1,  2, -2,  0,  2, -3, -1,  3, -3],
 [-2,  1, -1,  1,  1,  2, -2,  3,  3, -3, -1],
 [-2,  0, -1,  2, -1, -3, -2,  2,  1, -1, -2],
 [-2,  3,  1,  0,  0, -1, -2, -1, -3,  0,  0],
 [ 3, -2,  0,  0,  2, -1, -3, -3, -3, -3,  0],
 [ 2, -2, -2, -1,  2,  0, -1,  3,  1,  1, -3],
 [ 2,  1,  2,  0, -3, -3, -1, -3, -2,  2, -3],
 [ 3, -1,  1, -3,  1,  3,  0,  3,  3, -1,  1],
 [ 3, -3,  3, -2,  3,  1, -3,  3,  3, -2, -3],
 [-3, -3,  3,  3, -3, -2, -3,  3, -3,  3, -2],
 [ 1,  1,  0,  2, -1,  2,  1, -2,  2,  0,  3]]) #создание матрицы 
B=np.array([ 1, -1,  0, -1,  1,  0,  2, -3,  0,  0,  2]) #создаю матрицу столбца свободных членов 
det = np.linalg.det(A) 
if det==0:
    print ("Определитель матрицы равен нулю, система имеет бесконечно много решений")
else:
    a_inv = np.linalg.inv(A) 
X=np.dot(a_inv, B)
    
print ("Решение системы:\n", X)
```
